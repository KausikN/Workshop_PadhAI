# MP Neuron Class

```
class MPNeuron:
  
  def __init__(self):
    self.b = None
    
  def model(self, x):
    return (sum(x) >= self.b)
  
  def predict(self, X): # Predicts the value
    Y = []
    for x in X:
      result = self.model(x)
      Y.append(result)
    return np.array(Y)
  
  def fit(self, X, Y): # This finds the best value for b
    accuracy = {}
    
    for b in range(X.shape[1] + 1):
      self.b = b
      Y_pred = self.predict(X)
      #print(Y_pred)
      accuracy[b] = accuracy_score(Y_pred, Y)
      
    best_b = max(accuracy, key = accuracy.get) # finds max of accuracies nad returns KEY
    self.b = best_b
    
    print("Optimal value of b is", best_b)
    print("Highest accuracy is", accuracy[best_b])
```

# Perceptron Class

```
# We want to use perceptron on original data
# Not on the binarised data as perceptron works with real nos.
X_train = X_train.values # quick way to convert to numpy array
X_test = X_test.values
```

$y = 1, \mbox{if} \sum_i w_i x_i >= b$

$y = 0, \mbox{otherwise}$

```
class Perceptron:
  
  def __init__(self):
    self.w = None
    self.b = None
    
  def model(self, x):
    return 1 if(np.dot(self.w, x) >= self.b) else 0
    
  def predict(self, X):
    Y = []
    for x in X:
      result = self.model(x)
      Y.append(result)
    return np.array(Y)
    
  def fit(self, X, Y, epochs = 1, lr = 1):
    
    self.w = np.ones(X.shape[1]) # init w and b -- randomising is preferred
    seed = 2
    np.random.seed(seed)
    self.b = np.random.random()
    print(self.b,seed)
    
    accuracy = {}
    max_accuracy = 0
    
    wt_matrix = [] # Stores all weights as a matrix/list
    
    for i in range(epochs):
      for x, y in zip(X, Y):
        y_pred = self.model(x)
        if y == 1 and y_pred == 0:
          self.w = self.w + lr * x  # lr -- learning rate -- to control rate at which learning happens
          self.b = self.b - lr * 1
        elif y == 0 and y_pred == 1:
          self.w = self.w - lr * x
          self.b = self.b + lr * 1
          
      wt_matrix.append(self.w)
      accuracy[i] = accuracy_score(self.predict(X), Y)
      if(accuracy[i] > max_accuracy):
        max_accuracy = accuracy[i]
        chkptw = self.w # Checkpointing the parameters to get max accuracy
        chkptb = self.b
        
    self.w = chkptw
    self.b = chkptb
    
    print(max_accuracy)
      
    plt.plot(accuracy.values()) # Plotting accuracy
    plt.ylim([0, 1])
    plt.show()
    
    return np.array(wt_matrix)
```

```
perceptron = Perceptron()
```

**Hyper Parameters Used,**

  1. **Epochs**  -- Controls number of iterations through the data
  
  2. **Learning Rate** -- Controls the variations in the parameters -- can lead to high oscillations for high value of lr or low oscillations for low values of lr
  

**Hyper Parameter Tuning:**

  It is changing the hyper parameter values to get the maximum possible accuracy

```
 wt_matrix = perceptron.fit(X_train, Y_train, 100, 1) # Giving 2, 3 , ... epochs
 # Giving a low lr means your parameter dont oscillate too much
```

```
Y_pred_train = perceptron.predict(X_train)
print(accuracy_score(Y_pred_train, Y_train))
```

```
Y_pred_test = perceptron.predict(X_test)
print(accuracy_score(Y_pred_test, Y_test))
```

## Animating the Plot

```
%matplotlib inline
from matplotlib import animation, rc
from IPython.display import HTML
```

```
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim((0, wt_matrix.shape[1]))
ax.set_ylim((-15000, 25000))

line, = ax.plot([], [], lw=2)
```

```
# animation function. This is called sequentially
def animate(i):
  x = list(range(wt_matrix.shape[1]))
  y = wt_matrix[i,:]
  line.set_data(x, y)
  return (line,)
```

```
# call the animator. blit=True means only redraw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=100, interval=200, blit=True)
```
